### Partie 3 

#### Quelle est la part de commande qui prend plus de 15 minutes entre la commande et l'acceptation ?

```python
orders_dataset.head(5)
```

```python
orders_dataset.order_status.unique()
#On regarde les différents order status et on ne garde que ceux qui sont pertinents
```

```python
approved_status = ['delivered', 'invoiced', 'shipped','approved']
#On pourrait mettre tout le crochet à la place de "approved_status" ci-dessous, mais cette syntaxe
#est préférable pour permettre une modification plus simple en cas de besoin
```

```python
approved_orders = orders_dataset[orders_dataset.order_status.isin(approved_status)]
```

```python
approved_orders.order_approved_at.dtypes
#les colonnes timestamp ont perdu leur type, il y a donc besoin de les transformer
```

```python
approved_orders['purchase_time']=pd.to_datetime(approved_orders.order_purchase_timestamp)
approved_orders['approved_time'] = pd.to_datetime(approved_orders.order_approved_at)
```

```python
approved_orders.drop('order_purchase_timestamp',1)
```

```python
approved_orders.drop('order_approved_at',1)
# Ces deux étapes sont optionnelles, elles permettent de montrer comment supprimer une colonne et
# de gagner en lisibilité dans son tableau
```

```python
approved_orders['time_purchase_to_approve'] = approved_orders.approved_time - approved_orders.purchase_time
```

```python
approved_orders['time_purchase_to_approve'].head(10)
```

```python
too_long_approval = approved_orders[approved_orders.time_purchase_to_approve>'0 days 0:15:00.0']
```

```python
nb_approval_too_long = too_long_approval.order_id.count()
```

```python
total_number_of_approved_orders = approved_orders.order_id.count()
```

```python
part_of_too_long_orders = nb_approval_too_long/total_number_of_approved_orders
```

```python
part_of_too_long_orders
```

### Partie 4

#### Pour quelle catégorie de produit réaliser une page spécifique en priorité ?

```python
products_dataset.head(5)
```

```python
order_items_dataset.head(5)
```

```python
# Sélection des colonnes les plus pertinentes pour gagner en lisibilité (optionnel)
selected_product_cat_col= ['product_id','product_category_name']
selected_order_items_col = ['product_id','price', 'order_item_id']

product_cat = products_dataset[selected_product_cat_col]
order_it = order_items_dataset[selected_order_items_col]
```

```python
#calcu de la valeur totale de la commande
order_it['total_value']=order_it.order_item_id*order_it.price
```

```python
orders_with_product_cat = pd.merge(order_it, product_cat, how='left')
```

```python
orders_with_product_cat.head(10)
```

```python
aggregated_orders_by_cat = orders_with_product_cat.groupby('product_category_name').sum()
```

```python
aggregated_orders_by_cat = aggregated_orders_by_cat.reset_index()
```

```python
aggregated_orders_by_cat.sort_values('total_value', ascending=False).head(5)
```
